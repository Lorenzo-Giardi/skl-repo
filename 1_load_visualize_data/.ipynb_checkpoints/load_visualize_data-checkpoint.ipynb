### Change dtype of columns

```python
df = df.astype({'total_rooms': 'int32',
                'population': 'int32',
                'households': 'int32',
                'ocean_proximity': 'category'})
```

```python
df.info()
```

### Visualization

```python
%matplotlib inline
df.hist(bins=50, figsize=(15,10))
```

```python
df.plot(kind='scatter', x='longitude', y='latitude')
```

```python
df.plot(kind='scatter', x='longitude', y='latitude', alpha=0.15)
```

We can make a multidimensional plot leveraging color and other features:
* x-axis for longitude
* y-axis for latitude
* circle radius for population
* color for house prices

```python
df.plot(kind='scatter', x='longitude', y='latitude',
        s=df['population']/100, label='population',
        c='median_house_value', cmap=plt.get_cmap('jet'), colorbar=True,
         alpha=0.3, figsize = (10,7))
plt.legend()
```

### Correlations

```python
corr_matrix = df.corr()
```

```python
corr_matrix
```

```python
plt.figure(figsize=(10,8))
ax = sns.heatmap(
    corr_matrix, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
)
```
